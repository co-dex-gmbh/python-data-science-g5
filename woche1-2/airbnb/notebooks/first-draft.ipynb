```python
# Hypothese: Die NAs sind gro√üteils werte, die durch doppelte accounts zustande kommen
identify_na = df[df["host_identity_verified"].isna()][["NAME", "host name"]]
identify_not_na = df.dropna(subset=["host_identity_verified"])

identify_na.shape
```

```python
duplucated_host_identies = identify_not_na.merge(identify_na, on=["NAME", "host name"], how="inner")
duplucated_host_identies.head()
```

```python
# Einfachster Umgang mit diesem Wert: Ersetze NA durch "unknown"
# Was machen wir: Werte, die nicht doppelt sind, durch "unknown" ersetzen

df[(df["host_identity_verified"].isna()) &
   (df[["NAME", "host name"]].isin(
       duplucated_host_identies[["NAME", "host name"]]))
   ]
```
