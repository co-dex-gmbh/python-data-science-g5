```python
import pandas as pd

df = pd.read_csv("../data_set_original/penguins.csv")

df["male"] = df["sex"].apply(lambda x: 1 if x == "MALE" else 0)
df["female"] = df["sex"].apply(lambda x: 1 if x == "FEMALE" else 0)


print(f"male: {df["male"].sum()}")
print(f"female: {df["female"].sum()}")

def remove_na(self, col=[]):
    for c in col: 
        self.dropna(subset = c,inplace=True)
    df= self
    return df

pd.DataFrame.remove_na = remove_na

df.remove_na([c for c in df.columns])

```
