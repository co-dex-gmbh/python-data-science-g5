```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import os
```

```python
url = "https://www.bundestag.de/parlament/plenum/abstimmung/abstimmung"
output_folder = "bundestag_data"
```

```python
_url
```

```python

id = 1

for id in range(649, 948):
    print(id)
    _url = f"{url}?id={id}"
    body = requests.get(_url).text

    # Parse HTML content
    soup = BeautifulSoup(body, 'html.parser')

    # Get article title and clean it for folder name
    try:
        title = soup.find('h1', class_='bt-artikel__title').text.strip()
    except AttributeError:
        print(f"Failed to get title for id: {id}")
        continue
    folder_name = title
    folder_name = os.path.join(output_folder, folder_name)

    # Create folder if it doesn't exist
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    # Find all PDF links
    pdf_links = [a['href'] for a in soup.find_all('a', class_='dipLink')]

    # Download each PDF
    for link in pdf_links:
        filename = link.split('/')[-1]
        response = requests.get(link)

        if response.status_code == 200:
            filepath = os.path.join(folder_name, filename)
            with open(filepath, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded: {filename}")
        else:
            print(f"Failed to download: {filename}")
```

```python
number_of_pages = 95
next_button_class = "slick-next slick-arrow"
not_allowd = "/resource/blob/834414/"

url = "https://www.bundestag.de/parlament/plenum/abstimmung/liste"

body = requests.get(url).text
```

```python

soup = BeautifulSoup(body, 'html.parser')
rows = soup.find_all('tr')

results = []
for row in rows:
    # Get publication date
    date_cell = row.find('td', attrs={'data-th': 'Veröffentlichung'})
    if date_cell:
        date = date_cell.text.strip()
        
        # Get document info
        doc_cell = row.find('td', attrs={'data-th': 'Dokument'})
        if doc_cell:
            # Extract title from the strong tag
            title_elem = doc_cell.find('strong')
            if title_elem and title_elem.find('span'):
                title = title_elem.find('span').text.strip()
                
                # Find XLSX link
                xlsx_link = None
                links = doc_cell.find_all('a', class_='bt-link-dokument')
                for link in links:
                    if 'XLSX' in link.text:
                        xlsx_link = link['href']
                        
                if xlsx_link:
                    results.append({
                        'date': date,
                        'title': title,
                        'xlsx_url': xlsx_link
                    })

# Display results
for result in results:
    print(f"Date: {result['date']}")
    print(f"Title: {result['title']}")
    print(f"XLSX URL: {result['xlsx_url']}")
    print('-' * 50)
```
