```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random points for the inner class (Class 1)
np.random.seed(42)  # For reproducibility
num_points = 100
radius_inner = 1.0
inner_angles = np.random.uniform(0, 2 * np.pi, num_points)
inner_radii = np.random.uniform(0, radius_inner, num_points)

x_inner = inner_radii * np.cos(inner_angles)
y_inner = inner_radii * np.sin(inner_angles)

# Generate random points for the outer class (Class 2)
radius_outer_min = 1.5
radius_outer_max = 3.0
outer_angles = np.random.uniform(0, 2 * np.pi, num_points)
outer_radii = np.random.uniform(radius_outer_min, radius_outer_max, num_points)

x_outer = outer_radii * np.cos(outer_angles)
y_outer = outer_radii * np.sin(outer_angles)

# Plot the points
plt.figure(figsize=(8, 8))
plt.scatter(x_inner, y_inner, color='blue', label='Class 1 (Inner)', alpha=0.7)
plt.scatter(x_outer, y_outer, color='red', label='Class 2 (Outer)', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.title('Two Classes: Inner and Outer')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import plotly.express as px

z_inner = x_inner**2 + y_inner**2
z_outer = x_outer**2 + y_outer**2


df = pd.DataFrame({"x": x_inner, "y": y_inner, "z":z_inner})
df["class"] = "inner"

df_2 = pd.DataFrame({"x": x_outer, "y": y_outer, "z": z_outer})
df_2["class"] = "outer"

df = pd.concat([df, df_2])

fig = px.scatter_3d(df, "x", "y", "z", color="class")
fig.show()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

X = np.array([[-1, -1], [-2, -1], [1, 1], [2, 1], [2, 0.1]])
y = np.array([1, 1, 2, 2, 2])

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
svm = SVC(kernel="linear")
svm.fit(X, y)

w = svm.coef_[0]
b = svm.intercept_

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)

 
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    svm,
    X,
    plot_method="contour",
    colors="k",
    linestyles=["--", "-", "--"],
    levels=[-1, 0, 1],
    ax=ax,
)
```
